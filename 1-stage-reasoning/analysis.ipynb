```python
from google.colab import drive

drive.mount('/content/drive')
log_path = '/content/drive/MyDrive/csci2470final/log3.jsonl'
```

```python
import jsonlines
import matplotlib.pyplot as plt

data = []
with jsonlines.open(log_path) as reader:
    for obj in reader:
        data.append(obj)

print(data)
```

```python
tests_generated = []
valid_tests = []
passed_valid_tests = []
success = []
info = []
generator_reward = []
discriminator_reward = []
```

```python
for example in data:
  tests_generated.append(example["tests_generated"])
  valid_tests.append(example["valid_tests"])
  passed_valid_tests.append(example["passed_valid_tests"])
  success.append(example["success"])
  info.append(example["info"])
  generator_reward.append(example["generator_reward"])
  discriminator_reward.append(example["discriminator_reward"])
```

```python
x_axis = list(range(1, 101))
```

```python
plt.plot(x_axis, generator_reward)
plt.xlabel("Problem Number")
plt.ylabel("Generator Reward")
plt.title("Generator Rewards over Time")
plt.grid(True)
plt.show()
```

```python
plt.plot(x_axis, discriminator_reward)
plt.xlabel("Problem Number")
plt.ylabel("Discriminator Reward")
plt.title("Discriminator Rewards over Time")
plt.grid(True)
plt.show()
```

```python
plt.plot(x_axis, tests_generated, label="# Tests Generated")
plt.plot(x_axis, valid_tests, label="# Valid Tests")
plt.plot(x_axis, passed_valid_tests, label="# Passed Valid Tests")

plt.xlabel("Problem Number")
plt.ylabel("Number of Tests")
plt.title("Tests Data")
plt.legend()
plt.grid(True)
plt.show()
```
