```python
import importlib
import sandbox

importlib.reload(sandbox)
from sandbox import run_code_tests
```

```python
from data import load_problems

problems = load_problems("data/leetcode_formatted.json")
filtered_problems = [problem for problem in problems if "linked-list" not in problem.tags]
filtered_problems = [problem for problem in filtered_problems if "tree" not in problem.tags]
filtered_problems = [problem for problem in filtered_problems if "easy" == problem.difficulty]

len(filtered_problems)
```

```python
for i in range(1, 101):
    print("======================================================")
    problem = filtered_problems[i]
    print(f"PROBLEM: {problem.id}")
    ground_truth = problem.reference_solution
    ground_truth = ground_truth.replace("List", "list")
    ground_truth = ground_truth.replace("class Solution:\n  ", "")
    ground_truth = ground_truth.replace("self, ", "")

    generator_output = input("Enter generator output: ")
    generator_output = generator_output.replace("\\n", "\n")
    discriminator_output = input("Enter discriminator output: ")
    try:
        rewards, info = run_code_tests(generator_output, discriminator_output, ground_truth)
    except Exception as e:
        rewards = sandbox.Rewards(0, 0)
        info = {
            "tests_generated": 0,
            "valid_tests": 0,
            "passed_valid_tests": 0,
            "success": False,
            "info": str(e)
        }

    print("GENERATOR REWARD: ")
    print(rewards.generator_reward)
    print("DISCRIMINATOR REWARD: ")
    print(rewards.discriminator_reward)
    print(f"STATS:")
    print(info)

    info["problem"] = problem.id
    info["generator_reward"] = rewards.generator_reward
    info["discriminator_reward"] = rewards.discriminator_reward
    info["trial"] = i

    if i % 10 == 0:
        info["generator_output"] = generator_output
        info["discriminator_output"] = discriminator_output

    with open(log_path, "a") as f:
        f.write(json.dumps(info) + "\n")
```
